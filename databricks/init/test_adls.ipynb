```python
storage_account = "homepediadatalake" 

spark.conf.set(
    f"fs.azure.account.key.{storage_account}.dfs.core.windows.net",
    dbutils.secrets.get(scope="homepedia", key="adls-account-key")
)

print("✅ Connexion ADLS configurée")
```

```python
bronze_path = f"abfss://bronze@{storage_account}.dfs.core.windows.net/"

try:
    files = dbutils.fs.ls(bronze_path)
    print(f"✅ Container bronze accessible — {len(files)} éléments trouvés")
    for f in files:
        print(f"  {f.path}")
except Exception as e:
    print(f"❌ Erreur : {e}")
```
